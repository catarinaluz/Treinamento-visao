Roteiro da prática:
1. Ler a imagem original
2. Usar `cvtColor` para transformar a imagem para as escalas GRAY e HSV
3. Clonar a imagem cinza e criar uma imagem *preta*
4. Mostrar essas 4 imagens
5. Criar uma imagem branca
6. Redimensionar uma imagem
7. Mostrar essas 2 últimas imagens

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import random
from google.colab.patches import cv2_imshow
```

```python
image = cv.imread("semear.png")
cv2_imshow(image)
```

```python
img_gray = cv.cvtColor(image,cv.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

```python
img_hsv = cv.cvtColor(image,cv.COLOR_BGR2HSV)
cv2_imshow(img_hsv)
```

```python
img_black = img_gray.copy()
for row in range(0,img_black.shape[0]):
  for col in range(0,img_black.shape[1]):
    img_black[row,col] = 0.3*img_black[row,col]
```

```python
cv2_imshow(img_black)
```

```python
h = 300
w = 300
img_white = np.zeros([h,w,3],dtype=np.uint8)
img_white.fill(230)
cv2_imshow(img_white)
```

```python
img_resize = cv.resize(img_white,[500,500])
cv2_imshow(img_resize)
```
