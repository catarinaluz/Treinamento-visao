**Introdução a filtros**
* Average filter - média entre a soma do produto das posições sobrepostas e a substituição do valor central por essa média.
* Gaussian filter - kernels ponderados com diferentes valores, baseados em uma distribuição
gaussiana.
* Median filter - mediana dos valores da matriz dentro dessa área.
* Bilateral filter - retira o ruído da imagem
e preserva os contornos originais.

Roteiro da prática:

1. Criar funções para os quatro tipos de filtro
2. Importar a imagem e aplicar os filtros

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
!wget https://raw.githubusercontent.com/catarinaluz/Treinamento-visao/main/images/pratica3.png
```

```python
img = cv.imread("pratica3.png")
cv2_imshow(img)
```

```python
# Average filter
img_avg = cv.blur(img,ksize = (3,3)) 
cv2_imshow(img_avg)
```

```python
# Gaussian filter
img_gaus = cv.GaussianBlur(img,(3,3),cv.BORDER_DEFAULT)
cv2_imshow(img_gaus)
```

```python
# Median Filter
img_med = cv.medianBlur(img, ksize = 3)
cv2_imshow(img_med)
```

```python
# Bilateral filter
img_bil = cv.bilateralFilter(img,2,4,1)
cv2_imshow(img_bil)
```
